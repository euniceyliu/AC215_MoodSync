```python
import json
import pandas as pd
import os
```

```python
def extract_info(playlist_item):
    """Extracts song name/artist pairs, playlist title, and description"""
    playlist_dict = {}
    tracklist = playlist_item['tracks']
    # Extract artist name and track name
    artist_track_pairs = [f"{item['track_name']}: {item['artist_name']}" for item in tracklist]
    playlist_dict['songs'] = artist_track_pairs
    # Extract playlist title and description
    playlist_dict['title'] = playlist_item['name']
    playlist_dict['description'] = playlist_item['description']
    return playlist_dict
```

```python
directory = 'spotify_million_playlist_dataset/data/'
final_playlists = []

# Iterate over all files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.json'):
        file_path = os.path.join(directory, filename)
    with open(file_path, 'r') as file:
        data = json.load(file)
    playlists = data['playlists']
    playlists_w_desc = [playlist for playlist in playlists if "description" in playlist]
    final_playlists.extend(extract_info(playlist) for playlist in playlists_w_desc)
    if len(final_playlists) > 1500:
        break
```

```python
playlist_df = pd.DataFrame(final_playlists)
filtered_playlist_df = playlist_df[(playlist_df['description'].str.len() > 10) & (playlist_df['songs'].apply(len) < 20)]
filtered_playlist_df
```

```python
# Save dataframe
filtered_playlist_df.to_csv('finetune_data_df.csv', index=False)
```
